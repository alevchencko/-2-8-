```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
# Символьна змінна
x = sp.symbols('x')
# Функція f(x) = cos(2x)
f = sp.cos(2 * x)
# Знаходимо перші три похідні
f1 = sp.diff(f, x)
f2 = sp.diff(f1, x)
f3 = sp.diff(f2, x)
# Виводимо похідні
print("f'(x) =", f1)
print("f''(x) =", f2)
print("f'''(x) =", f3)
# Значення функції та похідних в точці x = 0
x0 = 0
f_x0 = f.subs(x, x0).evalf()
f1_x0 = f1.subs(x, x0).evalf()
f2_x0 = f2.subs(x, x0).evalf()
f3_x0 = f3.subs(x, x0).evalf()
# Поліном Тейлора третього ступеня
T = f_x0 + f1_x0*(x - x0) + (f2_x0/2)*(x - x0)**2 + (f3_x0/6)*(x - x0)**3
print("Поліном Тейлора 3-го порядку:", T.evalf())
# Перетворимо вирази на функції для обчислення на числових значеннях
f_lambdified = sp.lambdify(x, f, 'numpy')
T_lambdified = sp.lambdify(x, T, 'numpy')
# Будуємо графіки функції та її наближення
x_vals = np.linspace(-np.pi, np.pi, 1000)
f_vals = f_lambdified(x_vals)
T_vals = T_lambdified(x_vals)
plt.plot(x_vals, f_vals, label='f(x) = cos(2x)', color='blue')
plt.plot(x_vals, T_vals, label='Тейлор (3rd order)', color='red', linestyle='--')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Графік функції та наближення поліномом Тейлора')
plt.grid(True)
plt.show()
```
