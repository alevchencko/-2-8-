```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange
from numpy.polynomial.polynomial import Polynomial
# Вихідні дані
x_points = np.array([-1, 0, 1, 5])
y_points = np.array([11, 4, 7, -1])
# Побудова інтерполяційного багаточлена Лагранжа
poly = lagrange(x_points, y_points)
# Перетворення у форму полінома для зручного відображення коефіцієнтів
polynomial = Polynomial(poly).coef
# Точки, в яких потрібно знайти значення функції
x_values_to_evaluate = np.array([-0.5, 2, 3, 4.5])
y_values_approx = np.polyval(poly, x_values_to_evaluate)
# Побудова графіка
x_plot = np.linspace(min(x_points) - 1, max(x_points) + 1, 500)
y_plot = np.polyval(poly, x_plot)
plt.figure(figsize=(8, 6))
plt.plot(x_plot, y_plot, label=r'$L_n(x)$', color='blue')
plt.scatter(x_points, y_points, color='red', zorder=5, label='Задані точки')
plt.scatter(x_values_to_evaluate, y_values_approx, color='green', zorder=5, label='Інтерпольовані точки')
plt.title(r'Інтерполяція за допомогою полінома Лагранжа')
plt.xlabel(r'$x$')
plt.ylabel(r'$L_n(x)$')
plt.legend()
plt.grid(True)
plt.show()
print("Коефіцієнти полінома Лагранжа:", polynomial)
print("Інтерпольовані значення у заданих точках:", y_values_approx)
```
