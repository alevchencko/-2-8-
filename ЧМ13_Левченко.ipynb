# 1. **Метод Ейлера**

```python
import numpy as np
import matplotlib.pyplot as plt
# Визначаємо рівняння для задачі а
def f1(x, y):
    return x + np.cos(y / np.sqrt(11))
# Метод Ейлера
def euler_method(f, x0, y0, h, x_end):
    x_values = np.arange(x0, x_end + h, h)
    y_values = np.zeros(len(x_values))
    y_values[0] = y0
    for i in range(1, len(x_values)):
        y_values[i] = y_values[i-1] + h * f(x_values[i-1], y_values[i-1])
    return x_values, y_values
# Початкові умови для задачі а
x0_a, y0_a, x_end_a, h_a = 2.1, 2.5, 3.1, 0.1
# Рішення за методом Ейлера для задачі а
x_euler_a, y_euler_a = euler_method(f1, x0_a, y0_a, h_a, x_end_a)
# Візуалізація результату
plt.plot(x_euler_a, y_euler_a, 'r', label="Метод Ейлера")
plt.title("Задача А: Метод Ейлера")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

# 2. **Метод Ейлера-Коші**

```python
import numpy as np
import matplotlib.pyplot as plt
# Визначаємо рівняння для задачі б
def f2(x, y):
    return x + np.sin(y / np.e)
# Метод Ейлера-Коші
def euler_cauchy_method(f, x0, y0, h, x_end):
    x_values = np.arange(x0, x_end + h, h)
    y_values = np.zeros(len(x_values))
    y_values[0] = y0
    for i in range(1, len(x_values)):
        y_star = y_values[i-1] + h * f(x_values[i-1], y_values[i-1])
        y_values[i] = y_values[i-1] + (h / 2) * (f(x_values[i-1], y_values[i-1]) + f(x_values[i], y_star))
    return x_values, y_values
# Початкові умови для задачі б
x0_b, y0_b, x_end_b, h_b = 1.4, 2.5, 2.4, 0.1
# Рішення за методом Ейлера-Коші для задачі б
x_euler_cauchy_b, y_euler_cauchy_b = euler_cauchy_method(f2, x0_b, y0_b, h_b, x_end_b)
# Візуалізація результату
plt.plot(x_euler_cauchy_b, y_euler_cauchy_b, 'b', label="Метод Ейлера-Коші")
plt.title("Задача Б: Метод Ейлера-Коші")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
