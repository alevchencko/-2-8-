```python
import numpy as np
import matplotlib.pyplot as plt
#  
def f1(x, y):
    return np.sin(x + 2) - y - 1.5
def f2(x, y):
    return x + np.cos(y - 2) - 0.5
#  
x = np.linspace(-5, 5, 400)
y = np.linspace(-5, 5, 400)
X, Y = np.meshgrid(x, y)
Z1 = f1(X, Y)
Z2 = f2(X, Y)
plt.figure(figsize=(10, 8))
contour1 = plt.contour(X, Y, Z1, levels=[0], colors='blue')
contour2 = plt.contour(X, Y, Z2, levels=[0], colors='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('  ')
#   
plt.plot([], [], color='blue', label='sin(x + 2) - y = 1.5')
plt.plot([], [], color='red', label='x + cos(y - 2) = 0.5')
plt.legend()
plt.grid(True)
plt.show()
#   
def iterate_simple(x0, y0, tol=0.001, max_iter=100):
    x, y = x0, y0
    for i in range(max_iter):
        x_new = 0.5 - np.cos(y - 2)
        y_new = np.sin(x + 2) - 1.5
        if abs(x_new - x) < tol and abs(y_new - y) < tol:
            break
        x, y = x_new, y_new
    return x, y
#   ( )
x0, y0 = 0, 0
#  
root_x, root_y = iterate_simple(x0, y0)
print(f"': x = {root_x:.4f}, y = {root_y:.4f}")
```
